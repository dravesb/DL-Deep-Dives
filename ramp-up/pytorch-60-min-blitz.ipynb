### Intro to _torch.autograd_

_torch.autograd_ is PyTorch's automatic differentiation engine. 

NNs are just big nested functions $y = f_p(\ldots f_2(f_1(x))))$ where each funciton $f_i(\cdot)$ are parameterized by weights and biases we store in tensors. 

Training NNs requires

- Forward Propogation: computation of $\hat{y}$ for the given parameter set
- Backward Propogation: the process to adjust the paramters to make $\hat{e} = \hat{y} - y$ smaller


The foward pass is pretty straightforward. Backpropogation is more complex - it does this by traversing the computational graph backwards collecting the derivatives evaluated at the residual error for SGG updates. [See here for the math](https://www.youtube.com/watch?v=tIeHLnjs5U8).

For this tutorial, we're just going to show how it is used - not the math behind it.

```python
import torch
from torchvision.models import resnet18, ResNet18_Weights

model = resnet18(weights=ResNet18_Weights.DEFAULT)
data = torch.rand(1, 3, 64, 64)
labels = torch.rand(1, 1000)
```

```python
# forward pass
prediction = model(data)
prediction.shape
```

```python
# compute loss
loss = (prediction - labels).sum()
loss
```

```python
# backwards pass
loss.backward()
```

Notice how we called _backward()_ on the error tensor. Autograd then calculates and stores the gradients for each parameter in the .grad attribute.

To actually perform the SGD step, we need an optimizer to know what to do with all of those derivatives.

```python
optim = torch.optim.SGD(model.parameters(), lr=1e-2, momentum=0.9)
```

Finally we call _.step()_ for initiate gradient descnet. The optimizer is the one adjusting each parameter and by its gradient stored in .grad.

```python
optim.step()
```

So at a very high level we have

- _optim.zero\_grad()_  clear the gradeints from the previous loop
- _pred = model(data)_  foward pass 
- _loss = loss_fn(pred, true)_  compute loss
- _loss.backward()_  compute gradients for each parameter given loss
- _optim.step()_ use the optimizer to update parameters given gradients


There is a ton more to really understand for how back propogation works but lets leave this here for now.

### Differentiation in Autograd

How does autograd actually collect gradients? It finds all parameters with _requires_grad=True_ and that signals to autograd every operation on them should be tracked.

```python
import torch

a = torch.tensor([2.0, 3.0], requires_grad=True)
b = torch.tensor([6.0, 4.0], requires_grad=True)
```

```python
# comput a polynomial form of these tensors
Q = 3 * a**3 - b**2
```

if we treat (a,b) as parameters of a NN, and Q as the error we can take gradients of Q wrt to (a, b)

$\frac{\partial Q}{\partial a} = 9a^2$

$\frac{\partial Q}{\partial b} = -2b$





When we call _.backward()_ on _Q_ then autograd makes this computation and stores them in the .grad field for the respective tensors attributes a.grad and b.grad

```python
external = torch.tensor([1.0, 1.0])
Q.backward(gradient=external)
```

```python
print(f"Autograd: {a.grad}, Human: {9 * a**2}")
print(f"Autograd: {b.grad}, Human: {-2 * b}")
```
